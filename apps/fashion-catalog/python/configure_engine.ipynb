# Configure a Shaped engine to predict customer transactions

## Prerequisites
- A Shaped account
- 3 datasets uploaded to your account - items, users, transactions 

## Simple engine config

The simplest relevance engine is a semantic search model on articles of clothing. The basic fields are:
- name and version: defines engine metadata
- `data` config: defines the tables your engine will use
- `data.index` cnfig: defines how your table data is indexed for search

```yaml
version: v2
name: hm__clothing_semantic_search
data: 
    item_dataset: 
        name: hm__articles
        query: 
            select * from hm__articles limit 100 
    index: 
        search:
            item_fields: 
                - prod_name
                - prod_type_name
                - product_group_name
                - colour_group_name
                - department_name
```

## Add training configuration

Now let's train a retrieval policy for this. For simplicity, let's auto-tune a lightgbm model

```
version: v2
name: hm__clothing_semantic_search
data: 
    item_dataset: 
        name: hm__articles
        query: 
            select * from hm__articles limit 100 
    index: 
        search:
            item_fields: 
                - prod_name
                - prod_type_name
                - product_group_name
                - colour_group_name
                - department_name
training:
    models: 
        - policy_type: "auto-tune"
        - policies:
            - lightgbm
```
